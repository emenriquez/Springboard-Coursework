###    1. What test is appropriate for this problem? Does CLT apply?

For this data, I will be primarily looking at two columns - race and call. The sample size is sufficiently large, and we can independence in the data.

```python
# Number of duplicate entries in the data
print('Number of Duplicate Entries in the data: ',data.duplicated().sum())
```

The data is independent and sufficiently large so CLT applies here. The tests that we can use here are to look at the differences of proportions between callbacks for races labeled 'w' vs. 'b'.

###    2. What are the null and alternate hypotheses?

The null hypothesis of interest here is that race ('w' or 'b') has no effect on probability of callback. The alternate hypothesis would be that race influences the proportion of callbacks received.

###   3. Compute margin of error, confidence interval, and p-value.

```python
# Total entries
total_calls = data.shape[0]

print('Total Entries: ', total_calls)

# Race entries
total_b = data[data.race == 'b'].shape[0]
total_w = data[data.race == 'w'].shape[0]

frac_b = total_b / total_calls
frac_w = total_w / total_calls

print('# of black race entries: ',total_b)
print('# of white race entries: ',total_w)

print('fraction of black race entries: ',frac_b)
print('fraction of white race entries: ',frac_w)
```

there are an evenly distributed number of black and white entries in the dataset.

```python
# average success rate
calls_all = sum(data.call) / total_calls

print('Percent of successful callbacks for all entries: {0:0.2f} %'.format(100*calls_all))

# Success rates by race
calls_b = sum(data[data.race=='b'].call) / total_b
calls_w = sum(data[data.race=='w'].call) / total_w

print('Percent of successful callbacks for black race entries: {0:0.2f} %'.format(100*calls_b))
print('Percent of successful callbacks for white race entries: {0:0.2f} %'.format(100*calls_w))
```

```python
# Calculate Margin of Error
error = np.sqrt(calls_all*(1-calls_all)*(1/total_b + 1/total_w))
print('Margin of Error: {0:0.4f}'.format(error))
print('Confidence Interval for fraction of successful callbacks: between {0:0.4f} and {1:0.4f}'.format(calls_all-error, calls_all+error))
```

```python
# Compute Z score and corresponding p-value
z_score = (calls_w - calls_b)/error
p_value = 2*(1-stats.norm.cdf(z_score))

print('p value via proportion comparison test: {0:0.5f}'.format(p_value))
```

###    4. Write a story describing the statistical significance in the context or the original problem.

Although the number of race entries was equal for white and black, the data shows a successful rejection of the null hypothesis that race does not influence the rate of callbacks. White entries were much more likely, by an average of approximately 3.2%, to recieve a callback than those with resumes labeled as having black-sounding names.
